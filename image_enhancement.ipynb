```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to plot histogram
def plot_histogram(image, title, ax):
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])
    ax.plot(hist, color='black')
    ax.set_xlim([0, 256])
    ax.set_title(title)
    ax.grid()

# Read the image
image_path = r"C:\Users\ayush\Downloads\identifyHoles\enhancement\en5.jpg"
gray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Read the image in grayscale
colored_image = cv2.imread(image_path)  # Read the original image in color

# Apply Histogram Equalization
equalized_image = cv2.equalizeHist(gray_image)

# Apply AHE
clahe = cv2.createCLAHE(clipLimit=40.0, tileGridSize=(8, 8))
ahe_image = clahe.apply(gray_image)

# Apply CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_image = clahe.apply(gray_image)

# Apply Contrast Reversing
contrast_reversed_image = cv2.bitwise_not(gray_image)

# Apply Contrast Stretching
minmax_img = cv2.normalize(gray_image, None, 0, 255, cv2.NORM_MINMAX)

# Plot original image and its histogram
fig, axes = plt.subplots(6, 2, figsize=(12, 18))
axes[0, 0].imshow(gray_image, cmap='gray')
axes[0, 0].set_title('Original Image')
axes[0, 0].axis('off')
plot_histogram(gray_image, 'Original Histogram', axes[0, 1])

# Plot histogram equalized image and its histogram
axes[1, 0].imshow(equalized_image, cmap='gray')
axes[1, 0].set_title('Histogram Equalized Image')
axes[1, 0].axis('off')
plot_histogram(equalized_image, 'Equalized Histogram', axes[1, 1])

# Plot AHE image and its histogram
axes[2, 0].imshow(ahe_image, cmap='gray')
axes[2, 0].set_title('AHE Image')
axes[2, 0].axis('off')
plot_histogram(ahe_image, 'AHE Histogram', axes[2, 1])

# Plot CLAHE image and its histogram
axes[3, 0].imshow(clahe_image, cmap='gray')
axes[3, 0].set_title('CLAHE Image')
axes[3, 0].axis('off')
plot_histogram(clahe_image, 'CLAHE Histogram', axes[3, 1])

# Plot contrast reversed image and its histogram
axes[4, 0].imshow(contrast_reversed_image, cmap='gray')
axes[4, 0].set_title('Contrast Reversed Image')
axes[4, 0].axis('off')
plot_histogram(contrast_reversed_image, 'Contrast Reversed Histogram', axes[4, 1])

# Plot contrast stretched image and its histogram
axes[5, 0].imshow(minmax_img, cmap='gray')
axes[5, 0].set_title('Contrast Stretched Image')
axes[5, 0].axis('off')
plot_histogram(minmax_img, 'Contrast Stretched Histogram', axes[5, 1])

plt.tight_layout()
plt.show()

# Function to detect blobs
def detect_blobs(image, params):
    detector = cv2.SimpleBlobDetector_create(params)
    keypoints = detector.detect(image)
    return keypoints

# Parameters for small blobs
params_small = cv2.SimpleBlobDetector_Params()
params_small.minThreshold = 10
params_small.maxThreshold = 200
params_small.filterByArea = True
params_small.minArea = 1 
params_small.maxArea = 200
params_small.filterByCircularity = True
params_small.filterByConvexity = True
params_small.minConvexity = 0.2 
params_small.filterByInertia = False  
params_small.filterByColor = True
params_small.blobColor = 255 

# Detect small blobs on the equalized image
small_blobs = detect_blobs(median_blur, params_small)

# Draw circles on the contrast reversed image
for small_blob in small_blobs:
    small_centroid = (int(small_blob.pt[0]), int(small_blob.pt[1]))
    cv2.circle(contrast_reversed_image, small_centroid, 2, (0, 255, 0), 3)

# Display the contrast reversed image with detected blobs
cv2.imshow("Detected Blobs", contrast_reversed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Function to calculate the best-fit circle from given points
def best_fit_circle(points):
    A = np.c_[points[:, 0], points[:, 1], np.ones(points.shape[0])]
    f = np.square(points[:, 0]) + np.square(points[:, 1])
    C, _, _, _ = np.linalg.lstsq(A, f, rcond=None)
    xc = C[0] / 2
    yc = C[1] / 2
    r = np.sqrt(C[2] + xc**2 + yc**2)
    return xc, yc, r

# Extract blob coordinates from small_blobs
points = np.array([[small_blob.pt[0], small_blob.pt[1]] for small_blob in small_blobs])

if points.shape[0] >= 3:  # Ensure there are enough points to calculate the best-fit circle
    # Compute the best-fit circle
    xc, yc, r = best_fit_circle(points)

    # Draw the best-fit circle on the colored image
    cv2.circle(colored_image, (int(xc), int(yc)), int(r), (0, 255, 0), 3)

    large_centroid = (int(xc), int(yc))

    # Function to find if any small blob is close to a given point within a specified distance
    def is_blob_near_point(point, small_blobs, dis=1000):
        for small_blob in small_blobs:
            small_centroid = (int(small_blob.pt[0]), int(small_blob.pt[1]))
            distance = np.sqrt((small_centroid[0] - point[0]) ** 2 + (small_centroid[1] - point[1]) ** 2)
            if distance < dis:
                return True
        return False

    # Draw small blobs and extended lines
    for small_blob in small_blobs:
        small_centroid = (int(small_blob.pt[0]), int(small_blob.pt[1]))
        cv2.circle(colored_image, small_centroid, 2, (0, 255, 0), 3)
        cv2.line(colored_image, large_centroid, small_centroid, (0, 0, 255), 1)

        # Calculate direction vector
        direction = np.array(small_centroid) - np.array(large_centroid)
        direction = direction / np.linalg.norm(direction)  # Normalize the direction vector

        # Calculate intersection point with the other side of the large blob
        opposite_point = np.array(large_centroid) - direction * 1000  # Extend the line
        opposite_point = (int(opposite_point[0]), int(opposite_point[1]))

        # Draw the extended line
        cv2.line(colored_image, small_centroid, opposite_point, (255, 0, 0), 1)

        # Check a blob near the intersection point within a specified distance
        if is_blob_near_point(opposite_point, small_blobs):
            print("True")
        else:
            print("False")

    # Display the colored image with detected blobs and the best-fit circle
    cv2.imshow("Detected Blobs and Best-Fit Circle", colored_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    # Plotting for better visualization
    plt.figure(figsize=(10, 10))
    plt.imshow(cv2.cvtColor(colored_image, cv2.COLOR_BGR2RGB))
    plt.scatter(points[:, 0], points[:, 1], color='red', s=30)
    circle = plt.Circle((xc, yc), r, color='green', fill=False, linewidth=3)
    plt.gca().add_artist(circle)
    plt.gca().invert_yaxis()
    plt.title("Detected Blobs and Best-Fit Circle")
    plt.show()
else:
    print("Not enough points to compute the best-fit circle.")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the original color image
image_path = r"C:\Users\ayush\Downloads\identifyHoles\enhancement\en5.jpg"
color_image = cv2.imread(image_path)

# Convert the color image to grayscale
gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

# Apply different blurring filters
# Averaging Blur
average_blur = cv2.blur(gray_image, (5, 5))

# Gaussian Blur
gaussian_blur = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Median Blur
median_blur = cv2.medianBlur(gray_image, 5)

# Bilateral Filter
bilateral_blur = cv2.bilateralFilter(gray_image, 9, 75, 75)

# Display the original and blurred images
titles = ['Original Image', 'Averaging Blur', 'Gaussian Blur', 'Median Blur', 'Bilateral Blur']
images = [gray_image, average_blur, gaussian_blur, median_blur, bilateral_blur]

plt.figure(figsize=(15, 10))
for i in range(5):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
